# Outliers
- are the values which are outside the normal accepted range
- Handling Outliers:
- IMPUTE: update with mean,median or with capping
- Drop them: drop the row which has a Outlier
- What impact outliers can have?
- it can give biased results.
- for example, the mean of a column could be completly affected because of outliers.


```python
# how to find the outliers?
### answer?: look at quantiles values,boxplot visualization,by looking at the max or min values. for this descrive function could be useful

df["DURATION"].quantile(0.95)
```

```python
# Filtering the rows where duration is greater than 60

df[df['DURATION']>60]
```

```python
df.loc[7,"DURATION"] = 45  # so now 450 will get update with 45
```

```python
# capping of outliers: 120
arr=[]
for val in df["CALORIES BURNT"]:
  if val>400:
    arr.append(400)
  else:
    arr.append(val)

print(arr)
```

```python
df.groupby(["DURATION"]).agg({"CALORIES BURNT":"mean",
                              "PULSE":"max",
                              "DURATION":"size"}).rename(columns={"DURATION":"SIZE"})
```

# Concatenation of two dataframes

# Joining of two dataframes
- merge()


```python
df1 = pd.DataFrame({
    'productid': [101, 102, 103, 104],
    'product_name': ['Widget A', 'Widget B', 'Widget C', 'Widget D']
})

df2 = pd.DataFrame({
    'productid': [103, 104, 105, 106],
    'price': [25.50, 30.00, 15.75, 45.00]
})

print(df1)
print(df2)
```

```python
df1
```

```python
df2
```

# Inner Join

```python
df1.merge(df2, how="inner", on="productid")
```

# LEFT JOIN

```python
df1.merge(df2, how="left", on="productid")
```

# RIGHT JOIN

```python
df1.merge(df2, how="right", on="productid")
```

# OUTER JOIN

```python
df1.merge(df2, how="outer", on="productid")
```
