#sampling with replacement

- there is high chance that same value/row is selected again in the sample
- why?: as it is kept back(replaced) into the population

```python
df.sample(frac=0.3,replace=True,random_state=48).index.value_counts()
```

#sampling without replacement

- there is no chance of duplicate row Selection
- as it is removed from the population once it is selected

```python
df.sample(frac=0.3,replace=False,random_state=48).index.value_counts()
```
